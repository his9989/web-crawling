```python
import os
import sys
import urllib.request
import json
import csv
KakaoAK = "KakaoAK 581b8e8a9026504f3577a968ed7859f9"


from konlpy.tag import Twitter
from collections import Counter


params = {
    'sort': 'accuracy',
    'size': 50,
    'query': '커피'
}

def get_tags(text, ntags=50):
    spliter = Twitter()
    nouns = spliter.nouns(text)
    count = Counter(nouns)
    return_list = []
    for n, c in count.most_common(ntags):
        temp = {'tag': n, 'count': c}
        return_list.append(temp)
    return return_list

def kakaoblog(page):
    encText = urllib.parse.quote("스타벅스")
    url = "https://dapi.kakao.com/v2/search/blog?query=" + encText + "&page=" + str(page)
    print(url)
    request = urllib.request.Request(url)

    request.add_header("Authorization", KakaoAK)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    result_data = list()
    if (rescode == 200):
        response_body = response.read()
        result = response_body.decode('utf-8')
        json_test = json.loads(result)

        count = 0
        for temp in json_test["documents"]:
            print(temp['contents'])
            result_data.append(temp['contents'])
            count += 1

    else:
        print("Error Code:" + rescode)

    return result_data


with open('newsn.csv', 'w' , encoding='UTF-8') as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    csvout = csv.DictWriter(csvfile, ['description'])
    csvout.writeheader()

    str1 = ''

    for i in range(50):

            data = kakaoblog(i+1)


            for temp in data:
                str1 = str1+""+temp
                list_data = []
                result_data = temp
                list_data.append(result_data)
                writer.writerow(list_data)

    rdata = get_tags(str1,50)
    print(rdata)
```
